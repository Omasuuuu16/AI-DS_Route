```python
#iqr
df = pd.DataFrame(
    {
        "age": [1,2,3,20, 22, 23, 16, 19, 17, 18, 23 , 22, 25 , 50, 100 ]
    }
)
```

```python
df.describe()
```

```python
sns.boxplot(df)
```

```python
q1=df["age"].quantile(0.25)
q3=df["age"].quantile(0.75)
iqr=q3-q1
iqr
```

```python
threds=1
mask_outliers=(df["age"]<q1-iqr*threds)|(df["age"]>q3-iqr*threds)
```

```python
mask_outliers
```

```python
df[mask_outliers]
```

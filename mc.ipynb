```python
import imblearn
from imblearn.over_sampling import RandomOverSampler
```

```python
df=pd.DataFrame(
    {
        "city":["cairo","giza","alex","cairo","giza"],
        "temp":["hot","cold","veryhot","hot","cold"],
        "t":[1,1,1,0,0]
    }
)
```

```python
x=df.drop(columns=["t"])
y=df["t"]
```

```python
ROS=RandomOverSampler()
X,Y=ROS.fit_resample(x,y)
```

```python
x
```

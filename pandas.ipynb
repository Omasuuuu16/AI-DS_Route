2. Create a new column `Rating_Label` using `.apply()`:

   * "Excellent" if Rating  9
   * "Good" if 7  Rating < 9
   * "Poor" if Rating < 7
3. Create a new column `Decade` (e.g. 2010  "2010s") using `.apply()`.
4. Find the **longest movie title** using `.apply(len)`.
5. Using `.apply()`, calculate `Rating_Adjusted` where each rating is **+0.5 if Thriller, else unchanged**.
6. For each movie, calculate the **average rating** using `groupby` + `.apply()` instead of `.mean()`.
7. For each user, calculate how many "Excellent" ratings they gave using `.apply()`.

```python
movies = pd.DataFrame({
    "MovieID": [1, 2, 3, 4, 5, 6, 7],
    "Title": ["Inception", "Titanic", "Avatar", "La La Land", "Joker", "Avengers", "The Batman"],
    "Genre": ["Sci-Fi", "Romance", "Sci-Fi", "Musical", "Thriller", "Action", "Thriller"],
    "Year": [2010, 1997, 2009, 2016, 2019, 2012, 2022]
})

ratings = pd.DataFrame({
    "MovieID": [1, 1, 2, 2, 3, 4, 4, 5, 6, 6, 7, 7],
    "UserID": [101, 102, 101, 103, 102, 101, 104, 105, 101, 103, 102, 105],
    "Rating": [9, 8, 8, 7, 9, 8, 9, 7, 9, 8, 8, 9]
})
```

```python
x=pd.merge(movies,ratings,on="MovieID")
```

```python
x
```

```python
def z(r):
    if r>=9:
        return"excellent"
    elif r<9:
        return"Good"
    else:
        return"poor"
```

```python
ratings["Rating_label"]=ratings["Rating"].apply(z)
ratings
```

```python
df=sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.info()
```

```python
cat=["pclass","sex","embarked","class","who","adult_male","deck","embark_town","alive","alone"]
```

```python
rows,cols=5,2
fig,axes=plt.subplots(rows,cols,figsize= (15,7))

for indx,col in enumerate(cat):
    r,c= indx//cols, indx%cols
    ax=axes[r,c]
    sns.countplot(data=df,x=col,ax=ax)
    
plt.tight_layout()
plt.show()
```

```python
rows,cols=5,2
fig,axes=plt.subplots(rows,cols,figsize= (15,7))

for indx,col in enumerate(cat):
    r,c= indx//cols, indx%cols
    ax=axes[r,c]
    sns.countplot(data=df,x=col,ax=ax,hue="survived")
    
plt.tight_layout()
plt.show()
```

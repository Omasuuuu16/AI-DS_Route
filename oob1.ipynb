```python
class person:
    def __init__(self,name,age,gender):
        self.name=name
        self.age=age
        self.gender=gender
        self.welcome()
        
    def welcome(self):
        if self.gender=="female":
            print("welcome miss")
        elif self.gender=="male":
            print("welcome mr")
        else:
            print("welcome")
        
```

```python
class car:
    def __init__(self,brand,model,year):
        self.brand=brand
        self.model=model
        self.year=year
        self.start_engine()
        self.car_info()
        
    def start_engine(self):
        print("the engine of ",self.brand,self.model,"is now running")
        
    def car_info(self):
        print("the car is a ",self.year,self.brand,self.model)
```

```python
c=car("nissan","sunny",2000)
```

```python
c.brand
```

```python
class person:
    def __init__(self,name,age ):
        self.name=name
        self.age=age
        
    @classmethod
    def get_data(cls,data_str):
        name,age=data_str.split("-")
        return name,int(age)
```

```python
class person:
    not_allowed_names=["hey","hello","welcome","*"]
    def __init__(self,name):
        try:
            self.check(name)
            self.name=name
        except:
            print("not allowed")
    def check(self,name):
        if name not in person.not_allowed_names :
            print("welcome",name)
        else:
            raise ValueError("not allowed")
```

```python
p=person("*")
```

```python
class person:
    def __init__(self,name):
        self.name=name
    def __str__(self):
        return f"the name is {self.name}"
    
    
class op:
    def __init__(self,value):
        self.value=value
    def __add__(self, other):
        return op(self.value +other.value)
```

```python
x=op(5)
y=op(4)
```

```python
x.__add__(y)
```

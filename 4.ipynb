```python
def fun(s):
    v="aeiou"
    c=0
    for i in s:
        if i in v:
            print("vowels are ",i)
            c+=1
        else:
            print("there is no vowels")
    return c
```

```python
fun("omar")
```

```python
def check_pass(ps):
    if len(ps)<8:
        return False
    else:
        flag_d=False
        flag_u=False
    for c in ps:
        if c.isdigit():
            flag_d=True
        elif c.isupper():
            flag_u=True
    return flag_u and flag_d
    
```

```python
check_pass("Aomsdddedced1")
```

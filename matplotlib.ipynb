```python
fig=plt.figure(figsize=(15,7))

x=np.linspace(1,10,100)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.annotate("specialpoint",xy=(6,1),xytext=(7,1.5),arrowprops=dict(facecolor="red"))
plt.text(2,1.5,"point")
plt.xlabel("x",size=45,color="purple")
plt.ylabel("sin/cos",size=35,color="purple")
plt.title("GRAPH",size=55,color="purple")
plt.legend(loc="lower left")
plt.grid(linestyle="--",alpha=0.9)
plt.ylim(-2,5)
plt.xticks(rotation=45)
plt.show()
```

```python
fig,axes=plt.subplots(2,4,figsize=(15,7))
fig.suptitle("figure")
g=["A","B","C"]
h=[10,40,60]
axes[0,0].plot(x,np.sin(x))
axes[0,0].set_title("sin")
axes[0,1].plot(x,np.cos(x))
axes[0,2].barh(g,h)
axes[0,3].pie(h,labels=g)
axes[1,0].hist(h)
axes[1,1].scatter(g,h)
axes[1,2].plot(g,h)
axes[1,3].text(0.5,0.5,"plot")
```

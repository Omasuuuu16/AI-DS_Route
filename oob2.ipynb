```python
class address:
    def __init__(self,st,city,country):
        self.st=st
        self.city=city
        self.country=country
    
    def __str__(self):
        return f"the address is {self.st},{self.city},{self.country}"
```

```python
class person:
    def __init__(self,name,age,add_str):
        self.name=name
        self.age=age
        self.address=address(add_str.split(",")[0],add_str.split(",")[1],add_str.split(",")[2])
        
    def __str__(self):
        return f"the name {self.name} ,the age {self.age},the address {self.address}"
```

```python
class book:
    def __init__(self,title,author):
        self.title=title
        self.author=author
    def __str__(self):
        return f"the title is {self.title} and the author is {self.author}"
        
```

```python
c=book("the tree","ahmed")
```

```python
print(c)
```

```python
class library:
    def __init__(self):
        self.books=[]
        
    def ad(self,title,author):
        c=book(title,author)
        self.books.append(c)
        print("added books")
        
    def rb(self,title,author):
        for b in self.books:
            if b.title == title and b.author==author:
                self.books.remove(b)
        
    def pb(self):
        for b in self.books:
            print(b)
```

```python
l=library()
```

```python
l.ad("c","andrew")
```

```python
print(l)
```

```python
# multi level inheritance
class animal:
    def __init__(self,name):
        self.name=name
        
    def speak(self):
        print("speak")
```

```python
org=ocat("cat")
```

```python
class flyable:
    def fly(self):
        print("fly!")
    def speak(self):
        print("quack")
```

```python
class swimable:
    def swim(self):
        print("swim")
    def speak(self):
        print("quackle")
```

```python
class duck(flyable,swimable):
    pass
```

```python
d=duck()
```

```python
d.swim()
```

```python
d.fly()
```

```python
d.speak()
```

```python
#setter getter
class person:
    def __init__(self,name):
        self.__name=name
    def setter(self,name):
        self.__name=name
    def getter(self):
        return self.__name
```

```python
p=person("mai")
```

```python
p.getter()
```

```python
p.setter("ali")
```
